```python
# Load dataset
df = pd.read_csv("training_dataset.csv")
df['Date'] = pd.to_datetime(df['Date'])

# Define athlete and date range
athlete_player = 21
start_date = pd.to_datetime("7/10/2022", format='%d/%m/%Y')
end_date = pd.to_datetime("08/12/2022", format='%d/%m/%Y')

# Filter dataset
new_df = df[(df['Athlete'] == athlete_player) & (df['Date'] >= start_date) & (df['Date'] <= end_date)]

# Calculate average HRV, RHR, hours of sleep, sleep need, and sleep score
average_hrv = new_df['HRV'].mean()
average_rhr = new_df['RHR'].mean()
average_hours_of_sleep = new_df['Hours.of.Sleep'].mean()
average_sleep_need = new_df['Sleep.Need'].mean()
average_sleep_score = new_df['Sleep.Score'].mean()

# Print average values
print("Average HRV:", average_hrv)
print("Average RHR:", average_rhr)
print("Average Hours of Sleep:", average_hours_of_sleep)
print("Average Sleep Need:", average_sleep_need)
print("Average Sleep Score:", average_sleep_score)

neon_blue = '#00FFFF'
# Plot combined graph for deep sleep, awake, and light sleep analysis
plt.figure(figsize=(10, 6))
plt.plot(new_df['Date'], new_df['Deep.Sleep..hours.'], label='Deep Sleep', color=neon_blue)
plt.plot(new_df['Date'], new_df['Awake..hours.'], label='Awake', color= '#FFFF00')
plt.plot(new_df['Date'], new_df['Light.Sleep..hours.'], label='Light Sleep', color='#FFA500')
plt.xlabel('Date', color=neon_blue)
plt.ylabel('Hours', color=neon_blue)
plt.title('Deep Sleep, Awake, and Light Sleep Analysis', color=neon_blue)
plt.legend()
plt.xticks(rotation=45, color=neon_blue)
plt.yticks(color=neon_blue)

# Set color of the border lines to white
plt.gca().spines['top'].set_color('white')
plt.gca().spines['right'].set_color('white')
plt.gca().spines['bottom'].set_color('white')
plt.gca().spines['left'].set_color('white')

plt.tight_layout()
plt.savefig('sleep_analysis_plot.png', transparent=True)  # Save with transparent background
plt.show()
```
